```python
# Task 1: User Login Check
def login_check(user, pwd):
    # Fixed credentials
    username_db = "admin"
    password_db = "1234"

    if user == username_db and pwd == password_db:
        print("Login Successful")
    else:
        print("Invalid Credentials")

login_check("admin", "1234")
```

```python
# Task 2: Pass / Fail Analyzer
def analyze_marks(marks):
    pass_count = sum(1 for m in marks if m >= 50)
    fail_count = len(marks) - pass_count
    print(f"Total Pass: {pass_count}, Total Fail: {fail_count}")

analyze_marks([45, 78, 90, 33, 60])
```

```python
# Task 3: Simple Data Cleaner
def clean_names(names):
    cleaned = [name.strip().lower() for name in names]
    print(f"Cleaned Names: {cleaned}")

clean_names([" Alice ", "bob", " CHARLIE "])
```

```python
# Task 4: Message Length Analyzer
def analyze_messages(messages):
    for msg in messages:
        length = len(msg)
        flag = " [Long Message]" if length > 10 else ""
        print(f"Length: {length} | Message: {msg}{flag}")

analyze_messages(["Hi", "Welcome to the platform", "OK"])
```

```python
# Task 5: Error Message Detector
def count_errors(logs):
    error_count = logs.count("ERROR")
    print(f"Total Error Count: {error_count}")

count_errors(["INFO", "ERROR", "WARNING", "ERROR"])
```
