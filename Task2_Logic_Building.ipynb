```python
# Problem 1: Unique Words in a Sentence
def unique_word_analysis(sentence):
    words = sentence.lower().split()
    unique_words = set(words)
    print(f"Unique words count: {len(unique_words)}")
    print(f"Unique words: {unique_words}")

unique_word_analysis("python is easy and python is powerful")
```

```python
# Problem 2: Highest Salary from Employee Data
def find_highest_salary(employees):
    # Finding the key with the maximum value
    max_employee = max(employees, key=employees.get)
    print(f"Highest Salary: {max_employee} - {employees[max_employee]}")

employees = {"Ravi": 75000, "Anita": 68000, "Kiran": 72000}
find_highest_salary(employees)
```

```python
# Problem 3: Find Maximum and Minimum Values
def find_max_min(numbers):
    if not numbers:
        return
    maximum = minimum = numbers[0]
    for num in numbers:
        if num > maximum:
            maximum = num
        if num < minimum:
            minimum = num
    print(f"List: {numbers}")
    print(f"Max: {maximum}")
    print(f"Min: {minimum}")

# Function Call
nums = [45, 22, 89, 10, 66]
find_max_min(nums)
```

```python
# Problem 4: Count Products Above a Price Threshold
def count_expensive_products(prices, threshold=1000):
    count = 0
    for price in prices:
        if price > threshold:
            count += 1
    print(f"Products above {threshold}: {count}")

# Function Call
product_prices = [450, 1200, 899, 1500, 300]
count_expensive_products(product_prices)
```

```python
# Problem 5: Calculate Attendance Percentage
def calculate_attendance(attendance_list):
    present = attendance_list.count("P")
    total_days = len(attendance_list)
    percentage = (present / total_days) * 100
    print(f"Attendance Percentage: {percentage:.1f}")

# Function Call
attendance_data = ["P", "P", "A", "P", "P"]
calculate_attendance(attendance_data)
```

```python
# Problem 6: Remove Duplicate Phone Numbers
def remove_duplicate_phones(phone_list):
    unique_numbers = set(phone_list)
    print(f"Unique phone numbers: {unique_numbers}")

# Function Call
phone_numbers = [9876543210, 9123456789, 9876543210]
remove_duplicate_phones(phone_numbers)
```

```python
# Problem 7: Count Character Frequency
def character_frequency(text):
    freq = {}
    for char in text:
        freq[char] = freq.get(char, 0) + 1
    print(f"Character Frequency: {freq}")

# Function Call
character_frequency("python")
```

```python
# Problem 8: Convert List to Tuple
def convert_to_tuple(data_list):
    result = tuple(data_list)
    print(f"Tuple: {result}")

# Function Call
my_list = [10, 20, 30]
convert_to_tuple(my_list)
```

```python
# Problem 9: Check if a Key Exists in a Dictionary
def check_key_existence(data_dict, key_to_check):
    if key_to_check in data_dict:
        print(f"{key_to_check} exists")
    else:
        print(f"{key_to_check} does not exist")

# Function Call
sample_dict = {"Employee": "Bharath", "ID": 101}
check_key_existence(sample_dict, "Employee")
```

```python
# Problem 10: Calculate Average Marks
def calculate_average_marks(marks_list):
    if not marks_list:
        return
    average = sum(marks_list) / len(marks_list)
    print(f"Average Marks: {average}")

# Function Call
student_marks = [70, 80, 65, 75]
calculate_average_marks(student_marks)
```
