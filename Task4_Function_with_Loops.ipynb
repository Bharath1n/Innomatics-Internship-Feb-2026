```python
# Problem 1: Mobile Recharge Validation System

def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        try:
            amount = int(input("Enter recharge amount: "))

            if amount >= 50 and amount in valid_plans:
                print(f"Recharge of ₹{amount} Successful!")
                break
            else:
                print("Invalid Plan or Amount below ₹50. Please try again.")
        except ValueError:
            print("Please enter a valid numeric amount.")

# Function Call
validate_recharge()
```

```python
# Problem 2: Inventory Reorder Alert System

def check_inventory(inventory):
    for product, quantity in inventory.items():
        if quantity < 15:
            print(f"Reorder Alert: {product} (Stock: {quantity})")
        else:
            print(f"Stock OK: {product} ({quantity})")

# Input & Execution
stock_data = {"Laptops": 20, "Mice": 12, "Keyboards": 25, "Monitors": 8}
check_inventory(stock_data)
```

```python
# Problem 3: Student Result Processing System

def process_student_results(marks):
    if not marks:
        return "No marks provided"

    total = 0
    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return f"Average: {average:.2f} - Result: Pass"
    else:
        return f"Average: {average:.2f} - Result: Fail"

# Input & Execution
student_marks = [55, 60, 45, 70, 30]
print(process_student_results(student_marks))
```

```python
# Problem 4: Cab Fare Estimator with Retry Option

def calculate_fare(km, is_peak):
    base_fare = 50
    distance_fare = km * 12
    total = base_fare + distance_fare

    if is_peak:
        total *= 1.25  # 25% extra

    return total

def cab_app():
    while True:
        try:
            dist = float(input("Enter distance in km: "))
            peak = input("Is it peak hour? (yes/no): ").lower() == 'yes'

            fare = calculate_fare(dist, peak)
            print(f"Your estimated fare is: ₹{fare:.2f}")

            retry = input("Do you want to calculate again? (yes/no): ").lower()
            if retry != 'yes':
                break
        except ValueError:
            print("Invalid input. Please enter numbers for distance.")

# Function Call
cab_app()
```

```python
# Problem 5: Employee Attendance Eligibility Checker

def check_attendance_eligibility(attendance_list):
    if not attendance_list:
        return "No data"

    present_days = 0
    for status in attendance_list:
        if status == "P":
            present_days += 1

    percentage = (present_days / len(attendance_list)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# Input & Execution
attendance = ["P", "P", "A", "P", "P", "P", "A", "P"]
print(f"Status: {check_attendance_eligibility(attendance)}")
```

```python
# Problem 6: Password Strength Checker

def check_password_strength(password):
    if len(password) < 8:
        return "Weak: Minimum length should be 8 characters."

    has_digit = False
    has_special = False
    special_chars = "@#$"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak: Must contain at least one digit and one special character (@#$)."

# Input & Execution
print(check_password_strength("Python@123"))
print(check_password_strength("pass123"))
```
