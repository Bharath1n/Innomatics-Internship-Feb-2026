```python
# Problem 1 : Employee Performance Bonus Eligibility

def get_top_performers(employees):
    """Identifies the highest score and all employees who achieved it."""
    if not employees:
        return

    # Finding the maximum score
    max_score = max(employees.values())

    # Identifying all employees with the top score to handle ties
    top_performers = [name for name, score in employees.items() if score == max_score]

    # Display results
    print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {max_score})")

# Input & Execution
employees = {"Ravi": 92, "Anita": 88, "Kiran": 92, "Suresh": 85}
get_top_performers(employees)
```

```python
# Problem 2 : Search Query Keyword Analysis

import string

def analyze_search_keywords(query):
    """Counts keyword frequency, ignoring punctuation and case."""
    # Convert to lowercase and strip punctuation
    clean_query = query.lower().translate(str.maketrans('', '', string.punctuation))
    words = clean_query.split()

    # Count frequency and filter for keywords appearing more than once
    freq = {word: words.count(word) for word in set(words) if words.count(word) > 1}
    print(freq)

# Input & Execution
search_query = "Buy mobile phone buy phone online"
analyze_search_keywords(search_query)
```

```python
# Problem 3 : Sensor Data Validation

def validate_sensor_data(readings):
    """Filters valid (even) sensor readings and stores with their hour index."""
    # Store even readings as (hour_index, reading_value)
    valid_readings = [(i, val) for i, val in enumerate(readings) if val % 2 == 0]

    print("Valid Sensor Readings (Hour, Value):")
    print(valid_readings)

# Input & Execution
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
validate_sensor_data(sensor_readings)
```

```python
# Problem 4 : Email Domain Usage Analysis

def analyze_email_domains(emails):
    """Calculates the percentage usage of each email domain."""
    total_emails = len(emails)
    domain_counts = {}

    # Extract domains and count occurrences
    for email in emails:
        domain = email.split('@')[-1]
        domain_counts[domain] = domain_counts.get(domain, 0) + 1

    # Calculate and display percentages
    for domain, count in domain_counts.items():
        percentage = (count / total_emails) * 100
        print(f"{domain}: {percentage:.0f}%")

# Input & Execution
emails_list = ["ravi@gmail.com", "anita@yahoo.com", "kiran@gmail.com", "suresh@gmail.com", "meena@yahoo.com"]
analyze_email_domains(emails_list)
```

```python
# Problem 5 : Sales Spike Detection

def detect_sales_spikes(sales):
    """Detects days where sales are more than 30% above the average."""
    if not sales:
        return

    avg_sales = sum(sales) / len(sales)
    threshold = avg_sales * 1.3  # 30% threshold

    # Iterate and detect spikes
    for i, val in enumerate(sales, start=1):
        if val > threshold:
            print(f"Day {i}: {val}")

# Input & Execution
sales_data = [1200, 1500, 900, 2200, 1400, 3000]
detect_sales_spikes(sales_data)
```

```python
# Problem 6 : Duplicate User ID Detection

def detect_duplicate_users(user_ids):
    """Identifies and counts duplicate User IDs."""
    # Count occurrences
    counts = {uid: user_ids.count(uid) for uid in set(user_ids)}

    # Print only those that appear more than once
    for uid, count in counts.items():
        if count > 1:
            print(f"{uid} → {count} times")

# Input & Execution
user_ids_list = ["user1", "user2", "user1", "user3", "user1", "user3"]
detect_duplicate_users(user_ids_list)
```
