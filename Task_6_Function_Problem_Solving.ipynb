```python
# Problem 1: Smart Parking Lot Management System

def manage_parking(capacity, logs):
    parked_vehicles = 0

    for action in logs:
        if action == "IN":
            parked_vehicles += 1
        elif action == "OUT":
            parked_vehicles -= 1

    status = "Available" if parked_vehicles < capacity else "Full"

    if parked_vehicles > capacity:
        print("Alert: Parking Over Capacity!")

    print(f"Currently Parked Vehicles: {parked_vehicles}")
    print(f"Parking Status: {status}")

# Sample Execution
manage_parking(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

```python
# Problem 2: Online Food Delivery Time Estimator

def estimate_delivery_time(distance_km, traffic_level, weather):
    base_time_per_km = 5  # 5 minutes per km
    total_time = distance_km * base_time_per_km

    # Traffic Adjustments
    if traffic_level == "High":
        total_time += 10
    elif traffic_level == "Medium":
        total_time += 5

    # Weather Adjustments
    if weather == "Rainy":
        total_time += 5
    elif weather == "Stormy":
        total_time += 15

    print(f"Estimated Delivery Time: {total_time} minutes")

# Sample Execution
estimate_delivery_time(8, "High", "Rainy")
```

```python
# Problem 3: Movie Theatre Seat Occupancy Analyzer

def analyze_occupancy(total_seats, booked_seats_list):
    booked_count = len(booked_seats_list)
    occupancy_percentage = (booked_count / total_seats) * 100

    if occupancy_percentage == 100:
        status = "Housefull"
    elif occupancy_percentage >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"

    print(f"Occupancy: {occupancy_percentage:.0f}%")
    print(f"Show Status: {status}")

# Sample Execution (Representing 150 entries in list)
analyze_occupancy(200, [1] * 150)
```

```python
# Problem 4: Cloud Server Load Classification System

def classify_server_load(cpu_readings):
    average_cpu = sum(cpu_readings) / len(cpu_readings)

    if average_cpu < 50:
        status = "Normal"
    elif 50 <= average_cpu <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print(f"Average CPU Load: {average_cpu:.0f}%")
    print(f"Server Status: {status}")

# Sample Execution
classify_server_load([45, 60, 70, 85, 90])
```

```python
# Problem 5: Smart Classroom Resource Usage Monitor

def monitor_resources(usage_dict):
    overused = [res for res, hours in usage_dict.items() if hours > 8]
    energy_alert = "Yes" if overused else "No"

    print(f"Overused Resources: {', '.join(overused) if overused else 'None'}")
    print(f"Energy Alert: {energy_alert}")

# Sample Execution
resource_usage = {"Projector": 6, "AC": 9, "Lights": 4}
monitor_resources(resource_usage)
```

```python
# Problem 6: Online Event Registration Capacity Controller

def manage_registrations(capacity, total_registrations):
    if total_registrations > capacity:
        confirmed = capacity
        waitlisted = total_registrations - capacity
        status = "Closed"
    else:
        confirmed = total_registrations
        waitlisted = 0
        status = "Open"

    print(f"Confirmed Registrations: {confirmed}")
    print(f"Waitlisted Users: {waitlisted}")
    print(f"Registration Status: {status}")

# Sample Execution
manage_registrations(100, 105)
```
